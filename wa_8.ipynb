```python
import csv

def make_data_point(line, header):
    output = {}
    for index, header_item in enumerate(header):
        output[header_item] = line[index]
    return output

with open("denver-1.nov.csv") as inf:
    reader = csv.reader(inf)
    header = next(reader)
    all_data = []
    for line in reader:
        all_data.append(make_data_point(line, header))
```

```python
from collections import defaultdict

# this stores a dictionary mapping each precinct to a list of data
# from that precinct
byprecinct = defaultdict(list)  
for line in all_data:
    byprecinct[line["precinct"]].append(line)
    
byprecinct['423']
```

```python
data_byprecinct = defaultdict(list)

for point in all_data:
    data_byprecinct[point["precinct"]].append(point)
```

```python
arrestsby_precinct = {}

for precinct in data_byprecinct:
    data_forarrests = data_byprecinct[precinct]
    arrests = 0
    for stop in data_forarrests:
        if stop['arrest_made'] == 'TRUE':
            arrests += 1
    arrestsby_precinct[precinct] = arrests
    
arrestsby_precinct['value']
```

```python
highestValue = 0
for key,value in arrestsby_precinct.items():
    if(value > highestValue):
        highestValue = value
print(highestValue)
```

```python
arrest_list = []
for key,value in arrestsby_precinct.items():
    #print(value)
    if value > 0:
        arrest_list.append(value)
arrest_list
```

```python
import json

with open("output.json", "w") as of:
    of.write(json.dumps(arrestsby_precinct))
```

```python
with open("mostArrests.json", "w") as of:
    of.write(json.dumps(highestValue))
```

```python
with open("arrests.json", "w") as of:
    of.write(json.dumps(arrest_list))
```
